`````python
#Your code here

n_trees_list = np.floor(np.logspace(1, 3, num=10)).astype(int) #   

# 
train_mae_list = []
test_mae_list = []
time_list = []

for n_trees in n_trees_list:
    start_time = time.time()

    model = RandomForestRegressor(n_estimators=n_trees, random_state=42)
    model.fit(X_train, y_train)

    elapsed_time = time.time() - start_time
    time_list.append(elapsed_time)

    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    train_mae = mean_absolute_error(y_train, y_train_pred)
    test_mae = mean_absolute_error(y_test, y_test_pred)

    train_mae_list.append(train_mae)
    test_mae_list.append(test_mae)

    print(f"n_estimators={n_trees}, Time={elapsed_time:.2f}s, Train MAE={train_mae:.4f}, Test MAE={test_mae:.4f}")

#      
plt.figure(figsize=(10, 5))
plt.plot(n_trees_list, time_list, marker='o', linestyle='-', color='b')
plt.xscale('log')
plt.xlabel('  (n_estimators)')
plt.ylabel('  ()')
plt.title('     ')
plt.grid(True)
plt.show()

#  MAE   
plt.figure(figsize=(10, 5))
plt.plot(n_trees_list, train_mae_list, marker='o', linestyle='-', color='r', label='MAE   ')
plt.plot(n_trees_list, test_mae_list, marker='o', linestyle='-', color='g', label='MAE   ')
plt.xscale('log')
plt.xlabel('  (n_estimators)')
plt.ylabel('MAE')
plt.title(' MAE   ')
plt.legend()
plt.grid(True)
plt.show()
`````

`````python
# your code here

from sklearn.linear_model import Lasso
from sklearn.preprocessing import StandardScaler

data = fetch_california_housing()
X, y = data.data, data.target
feature_names = data.feature_names

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42) #  

#  
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

alphas = [0.1, 0.2, 0.5]

for alpha in alphas:
    model = Lasso(alpha=alpha, random_state=42)
    model.fit(X_train_scaled, y_train)

    y_pred = model.predict(X_test_scaled)

    mae = mean_absolute_error(y_test, y_pred)

    #  
    nonzero_coefs = np.sum(model.coef_ != 0)
    # ,   
    zeroed_features = [feature_names[i] for i in range(len(feature_names)) if model.coef_[i] == 0]

    print(f"\n alpha={alpha}:")
    print(f"MAE  : {mae:.4f}")
    print(f"  : {nonzero_coefs}")
    print(f" : {zeroed_features}")
`````

`````python
def select_best_features(X_train, X_test, y_train, y_test, max_features=3):
    best_features = []
    remaining_features = list(range(X_train.shape[1]))

    for _ in range(max_features):
        best_mae = float('inf')
        best_feature = None

        for feature in remaining_features:
            current_features = best_features + [feature]
            model = DecisionTreeRegressor(random_state=42)
            model.fit(X_train[:, current_features], y_train)
            y_pred = model.predict(X_test[:, current_features])
            mae = mean_absolute_error(y_test, y_pred)

            if mae < best_mae:
                best_mae = mae
                best_feature = feature

        best_features.append(best_feature)
        remaining_features.remove(best_feature)

    return best_features

data = fetch_california_housing()
X, y = data.data, data.target
feature_names = data.feature_names

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

best_features = select_best_features(X_train_scaled, X_test_scaled, y_train, y_test, max_features=3)
print(" :", [feature_names[i] for i in best_features])

lr_all = LinearRegression()
lr_all.fit(X_train_scaled, y_train)
y_pred_all = lr_all.predict(X_test_scaled)
mae_all = mean_absolute_error(y_test, y_pred_all)

X_train_selected = X_train_scaled[:, best_features]
X_test_selected = X_test_scaled[:, best_features]

lr_selected = LinearRegression()
lr_selected.fit(X_train_selected, y_train)
y_pred_selected = lr_selected.predict(X_test_selected)
mae_selected = mean_absolute_error(y_test, y_pred_selected)

print("MAE ( ):", mae_all)
print("MAE ( ):", mae_selected)
`````
