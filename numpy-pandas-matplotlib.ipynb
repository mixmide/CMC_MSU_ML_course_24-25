<font color='CornflowerBlue'>**Matplotlib**</font> - основная библиотека визуализации двумерных данных на языке python.

Рисунок создается с помощью последовательного вызова функций из модуля **pyplot** (создание рисунка, задание области рисования, построение графиков в области рисования, украшение графика метками). При этом различные графические элементы рисунка (линии, подписи и т.д.) перекрывают предыдущие, если они имеют общую площадь.

Основные типы графиков:
- <font color='DarkSeaGreen'>**line plot**</font> — обычный поточечный график, которым можно изображать кривые, соединяющие последовательные точки;
- <font color='DarkSeaGreen'>**histogram**</font> — гистограмма, показывающая распределение некоторое величины;
- <font color='DarkSeaGreen'>**scatter plot**</font> — диаграмма рассеивания, показывающая взаимосвязь двух величин;
- <font color='DarkSeaGreen'>**bar plot**</font> — столбчатая диаграмма, показывающий взаимосвязь количественной величины от категориальной.

#### <font color='DarkOrange'>**Задача 1 [0.3 + 0.7 баллов]**</font>

Вам дан двумерный непустой массив X размера n x m вещественных чисел.
Создайте на его основе массив состоящий: из каждого 4го (начиная с 0го и до конца) элемента по оси размерности n (axis=0), начиная с 120 до 500 (не включитально) с шагом 5 элементов по оси размерности m (axis=1).

Гарантируется, что m >= 500

 **Пример:**

      Input: X = [[1, 2, 3, ..., 500], [1, 2, 3, ..., 500], [1, 2, 3, ..., 500], [1, 2, 3, ..., 500], [1, 2, 3, ..., 500]]
      Output: [[121, 126, 131, ..., 496]]

```python
from typing import List
from copy import deepcopy
```

```python
# без numpy
def get_part_of_array(X: List[List[float]]) -> List[List[float]]:
    # определение размерности
    n = len(X)
    m = len(X[0])

    result = [] # результирующий массив
    for i in range(0, n, 4): # сначала проход по оси 0
        help = [] # вспомогательный массив
        for j in range(120, 500, 5): # проход по оси 1
            help.append(X[i][j])
        result.append(help)

    return result
```

```python
# векторизированная реализация
def get_part_of_array(X: np.ndarray) -> np.ndarray:
    rows = X[::4] # просмотр по оси 0
    result = rows[:, 120:500:5] # просмотр по оси 1 с учетом просмотренной оси 0
    return result
```

#### <font color='DarkOrange'>**Задача 2 [0.3 + 0.7 баллов]**</font>

Вам дан двумерный непустой массив X целых чисел размера n x m.

Посчитайте сумму неотрицательных элементов на главной диагонали прямоугольной матрицы X. Если неотрицательных элементов на диагонали нет, то верните -1.

  **Пример:**
  
      Input: X = np.array([[-1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]])
      Output: 3

```python
# без numpy
def sum_non_neg_diag(X: List[List[int]]) -> int:
    # определение размерности
    n = len(X)
    m = len(X[0])

    total = 0
    no_error = False

    for i in range(min(n, m)):
        if X[i][i] >= 0:
            no_error = True
            total += X[i][i]

    return total if no_error else -1
```

```python
# векторизированная реализация
def sum_non_neg_diag(X: np.ndarray) -> int:
    diagonal = np.diag(X) # извлекаем главную диагональ
    new_diagonal = diagonal[diagonal >= 0] # фильтрация

    # проверка на пустоту
    if new_diagonal.size == 0:
        return -1

    return new_diagonal.sum()
```

#### <font color='DarkOrange'>**Задача 3 [0.3 + 0.7 баллов]**</font>

Дан двумерный непустой массив X вещественных чисел размера n x m. Считая для каждого столбца отдельно, замените в копии массива те значения, которые больше 1.5 средних по столбцу или меньше 1/4 среднего по столбцу на число -1

<font color='LightSteelBlue'>**Подсказка**</font>

Возможно пригодится функция deepcopy из модуля copy

```python
# без numpy
def replace_values(X: List[List[float]]) -> List[List[float]]:
    # определение размерности
    n = len(X)
    m = len(X[0])
    # копия
    X2 = deepcopy(X)

    for j in range(m):
        average = sum([X[i][j] for i in range(n)]) / n   # среднее значение по столбцу
        for i in range(n):
            if (X[i][j] > (1.5 * average)) or (X[i][j] < (0.25 * average)):
                X2[i][j] = -1

    return X2
```

```python
# векторизированная реализация
def replace_values(X: np.ndarray) -> np.ndarray:
    # копия
    X2 = np.copy(X)
    average = np.mean(X2, axis = 0) # срредние значения по каждому столбцу
    # замена необходимых значений
    X2[(X > (1.5 * average)) | (X < (0.25 * average))] = -1

    return X2
```
