#### <font color='DarkOrange'>** 1 [0.3 + 0.7 ]**</font>

     X  n x m  .
     :   4 (  0   )     n (axis=0),   120  500 ( )   5     m (axis=1).

,  m >= 500

 **:**

      Input: X = [[1, 2, 3, ..., 500], [1, 2, 3, ..., 500], [1, 2, 3, ..., 500], [1, 2, 3, ..., 500], [1, 2, 3, ..., 500]]
      Output: [[121, 126, 131, ..., 496]]

```python
from typing import List
from copy import deepcopy
```

```python
#  numpy
def get_part_of_array(X: List[List[float]]) -> List[List[float]]:
    #  
    n = len(X)
    m = len(X[0])

    result = [] #  
    for i in range(0, n, 4): #     0
        help = [] #  
        for j in range(120, 500, 5): #    1
            help.append(X[i][j])
        result.append(help)

    return result
```

```python
#  
def get_part_of_array(X: np.ndarray) -> np.ndarray:
    rows = X[::4] #    0
    result = rows[:, 120:500:5] #    1     0
    return result
```

#### <font color='DarkOrange'>** 2 [0.3 + 0.7 ]**</font>

     X    n x m.

         X.      ,   -1.

  **:**
  
      Input: X = np.array([[-1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]])
      Output: 3

```python
#  numpy
def sum_non_neg_diag(X: List[List[int]]) -> int:
    #  
    n = len(X)
    m = len(X[0])

    total = 0
    no_error = False

    for i in range(min(n, m)):
        if X[i][i] >= 0:
            no_error = True
            total += X[i][i]

    return total if no_error else -1
```

```python
#  
def sum_non_neg_diag(X: np.ndarray) -> int:
    diagonal = np.diag(X) #   
    new_diagonal = diagonal[diagonal >= 0] # 

    #   
    if new_diagonal.size == 0:
        return -1

    return new_diagonal.sum()
```

#### <font color='DarkOrange'>** 3 [0.3 + 0.7 ]**</font>

    X    n x m.     ,      ,   1.5      1/4      -1

<font color='LightSteelBlue'>****</font>

   deepcopy   copy

```python
#  numpy
def replace_values(X: List[List[float]]) -> List[List[float]]:
    #  
    n = len(X)
    m = len(X[0])
    # 
    X2 = deepcopy(X)

    for j in range(m):
        average = sum([X[i][j] for i in range(n)]) / n   #    
        for i in range(n):
            if (X[i][j] > (1.5 * average)) or (X[i][j] < (0.25 * average)):
                X2[i][j] = -1

    return X2
```

```python
#  
def replace_values(X: np.ndarray) -> np.ndarray:
    # 
    X2 = np.copy(X)
    average = np.mean(X2, axis = 0) #     
    #   
    X2[(X > (1.5 * average)) | (X < (0.25 * average))] = -1

    return X2
```
